# EXTRACCION DE DATOS:

```python
# Set start and end dates for the price data

# Se establece la fecha de inicio para los datos de precios
start_date = '2020-01-01'
print(start_date)
print(type( start_date ))
```

```python
# Set end dates for the price data
# Establecer fechas finalización para los datos de precios

end_date = pd.Timestamp.today().strftime('%Y-%m-%d')

print(end_date)
print(type( end_date ))
```

```python
# Retrieve the price data from Yahoo Finance
# Se obtienen los datos de precios de Yahoo Finance

df = yf.download('BVN', start=start_date, end=end_date)
df
```

```python
## DATAFRAME de Pandas
print(type( df ))
display(len( df )) # shape len
display( df.info() )
display(df.describe())
display(df.sample(2))
display( df ) # .head(3) .tail(2) [:3]
```

# **TRANSFORMACION Y/O LIMPIEZA - PRE-PROCESAMIENTO:** ETL

```python
df['Date'] = df.index
df.sample(2)
```

```python
display( df.info() )
```

CREACION DE VARIABLES CALCULADAS x(1) y x(2) (los inputs) para formar nuestra MATRIZ "X".

Create predictor variables

VARIABLE CALCULADA x(1): Rango de la Vela

```python
df['Open-Close'] = df.Open - df.Close
df.sample(2)
```

VARIABLE CALCULADA x(2): Rango de la Aguja

```python
df['High-Low'] = df.High - df.Low
df.sample(2)
```

# Formar nuestra MATRIZ "X":

```python
# Store all predictor variables in a variable X
X = df[['Open-Close', 'High-Low']]
X.sample(2)
```

# Formar nuestro campo Retornos % Reales:

```python
df['Return'] =  df.Close.pct_change()
df
```

# Formar nuestro campo Etiqueta Binaria "Trend" (Tendencia) Real (1,0), corrida un lugar hacia arriba:

```python
# Target variables
y_real = np.where(df['Close'].shift(-1) > df['Close'], 1, 0)
```

## PREDICT (generar valores ajustados y/o predicción)

```python
df["y_real_total"] = y_real_total
df
```

```python
df['y_pred_total'] = Modelo_SVC1.predict(X)
df.head(3)
```

```python
# Calculate strategy returns
df['Strategy_Return'] = df.Return *df.y_pred_total.shift(1)
df['Strategy_Return']
```

```python
# Calculate Cumulutive returns
df['Cum_Ret'] = df['Return'].cumsum()
df
```

```python
# Plot Strategy Cumulative returns
df['Cum_Strategy'] = df['Strategy_Return'].cumsum()
df
```

```python

import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(df['Cum_Ret'],color='red')
plt.plot(df['Cum_Strategy'],color='blue')
```
